```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
with open('api_json/base_api.json', 'r') as f:
    data = json.load(f)
```

```python
print(data)
```

```python
tokenizer = AutoTokenizer.from_pretrained("stabilityai/stablelm-2-1_6b")
model = AutoModelForCausalLM.from_pretrained(
  "stabilityai/stablelm-2-1_6b",
  torch_dtype="auto",
)
```

```python
model.cuda()
```

```python
for entry in data:
    text = entry["description"]

    # Check if the text is empty and skip if so
    if not text.strip():
        print(f"Skipping entry with empty description: {entry['title']}")
        continue

    inputs = tokenizer(text, return_tensors="pt").to(model.device)

    # Ensure input_ids are of type long
    inputs['input_ids'] = inputs['input_ids'].long()

    tokens = model.generate(
        **inputs,
        max_new_tokens=64,
        temperature=0.70,
        top_p=0.95,
        do_sample=True,
    )
    generated_text = tokenizer.decode(tokens[0], skip_special_tokens=True)
    print(f"Title: {entry['title']}\nOriginal Description: {text}\nGenerated Text: {generated_text}\n")
```
